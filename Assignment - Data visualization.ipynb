```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
linkedin_df = pd.read_csv('Linkedin_data.csv')
linkedin_df.head()
```

```python
linkedin_df.shape
```

```python
linkedin_df.info()
```

```python
linkedin_df.describe(include='all').T
```

```python
linkedin_df.isnull().sum()
```

```python
linkedin_df.drop(['job_details', 'details_id'], axis=1, inplace=True)
linkedin_df
```

```python
linkedin_df.columns
```

```python
linkedin_df_1 = linkedin_df.copy()
```

```python
linkedin_df_1.dtypes
```

```python
linkedin_df_1.total_applicants.isnull().sum()
```

```python
linkedin_df_1.fillna(linkedin_df_1.total_applicants.mean(), inplace=True)
```

```python
linkedin_df_1.total_applicants
```

```python
for i in linkedin_df_1.columns:
    print(f"{i}: {linkedin_df_1[i].nunique()}")
```

```python
cat_cols = []
num_cols = []
for i in linkedin_df_1.columns:
    if linkedin_df_1.dtypes[i] == 'object':
        cat_cols.append(i)
    else:
        num_cols.append(i)

print(f"Categorical columns: {cat_cols}")
print(f"Numerical columns: {num_cols}")
```

```python
for i in cat_cols:
    print(f"{i}:- \n{linkedin_df_1[i].value_counts(normalize=True)*100}")
    print("---------------------------------------------\n---------------------------------------------")
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle('Distribution of work type', fontsize=16, fontweight='bold')
plt.bar(linkedin_df_1['work_type'].value_counts().sort_values(ascending=False).index, linkedin_df_1['work_type'].value_counts().values)
plt.xlabel('Work_type')
plt.ylabel('Counts')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle('Distribution of involvement', fontsize=16, fontweight='bold')
plt.barh(linkedin_df_1['involvement'].value_counts().index, linkedin_df_1['involvement'].value_counts().values)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle('Distribution of level', fontsize=16, fontweight='bold')
plt.bar(linkedin_df_1['level'].value_counts().index, linkedin_df_1['level'].value_counts().values)
plt.show()
```

```python
top10_designation_categories = linkedin_df_1.designation.value_counts().nlargest(10).index
top10_designation_filter = linkedin_df_1[linkedin_df_1.designation.isin(top10_designation_categories)]
```

```python
top10_designation_filter
```

```python
cross_tab = pd.crosstab(linkedin_df_1['work_type'], linkedin_df_1['involvement'], normalize='columns') * 100
cross_tab
sns.heatmap(cross_tab, annot = True)
plt.show()
```

```python
cross_tab1 = pd.crosstab(linkedin_df_1['work_type'], linkedin_df_1['level'], normalize='columns') * 100
cross_tab1
sns.heatmap(cross_tab1, annot = True)
plt.show()
```

```python
sns.barplot(x='industry',y='employees_count',data=linkedin_df_1[linkedin_df_1.industry!='Not Avilable'].head(100).sort_values('employees_count',ascending=False))
plt.xticks(rotation=90)
plt.title('Largest Number of Employees Count')
plt.show()
```
